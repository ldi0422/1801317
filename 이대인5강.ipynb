```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
import matplotlib.pyplot as plt
```

```python
vals = iris['sepal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for sepal length")
plt.show()
```

```python
vals = iris['petal_length'].values
sns.kdeplot(vals)
plt.title("kdeplot for petal length")
plt.show()
```

```python
vals = iris['petal_length'].values
sns.distplot(vals, kde=True, rug=True)
plt.title("distplot for petal lenght")
plt.show()
```

```python
sns.displot(vals, kde=True, rug=True)
plt.title("displot for petal lenngth")
plt.show()
```

```python
sns.rugplot(vals)
```

```python
sns.distplot(vals)
```

```python
sns.histplot(vals, kde=True)
plt.title("histplot for petal length")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.countplot(x="adult_male",data=titanic)
plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.countplot(x="day",data=tips)
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris, kind='scatter')
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris, hue='species', kind='kde')
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris, hue='species', markers= ['o','s','d'])
plt.show()
```

```python
titanic_size=titanic.pivot_table(index="class",columns="sex",aggfunc="size")
titanic_size
```

```python
sns.heatmap(titanic_size)
plt.title("Heatmap")
plt.show()
```

```python
sns.heatmap(titanic_size, annot=True, fmt='d')
plt.title("Heatmap")
plt.show()
```

```python
sns.heatmap(titanic_size, cmap=sns.color_palette("PuBu"), annot=True, fmt='d')
plt.title("Heatmap")
plt.show()
```
